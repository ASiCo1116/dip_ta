# Problem 1
### (1) Show the balloon in the 'gray' color map.
### (2) Show the balloon in the 'pink' color map.
### (3) Show the balloon in the 'cool' color map.
### (4) Show the coins in the 'gray' color map.

```python
from skimage import io, data, img_as_float #import the io, data module from scikit-image
import matplotlib.pyplot as plt #import the pyplot module to load and save the image
import numpy as np
```

```python
# (1)
balloon = io.imread('balloon.jpg', as_gray=True)
plt.imshow(balloon, cmap='gray')
plt.show()

# (2)
plt.imshow(balloon, cmap='pink')
plt.show()

# (3)
plt.imshow(balloon, cmap='cool')
plt.show()

# (4)
coins = data.coins()
plt.imshow(coins, cmap='gray')
plt.show()
```

# Problem 2
### (1) Show the mm chocolate in the folder.
### (2) Increase the brightness of the mm chocolate by addition.
### (3) Decrease the brightness of the mm chocolate by subtraction.
### (4) Increase the brightness of the mm chocolate by multiplication.
### (5) Decrease the brightness of the mm chocolate by division.

```python
mm = io.imread('mm.jpg', as_gray=True)
plt.imshow(mm, cmap='gray')
plt.show()

mm_bright = mm + 0.3
plt.imshow(mm_bright, cmap='gray', vmax=1., vmin=0.)
plt.show()

mm_dark = mm - 0.3
plt.imshow(mm_dark, cmap='gray', vmax=1., vmin=0.)
plt.show()

mm_bright_m = mm * 1.5
plt.imshow(mm_bright_m, cmap='gray', vmax=1., vmin=0.)
plt.show()

mm_dark_d = mm / 1.5
plt.imshow(mm_dark_d, cmap='gray', vmax=1., vmin=0.)
plt.show()
```

# Problem 3
### (1) Show the rice after being binarized with threshold = 0.7
### (1) Show the rice after being binarized with threshold = 0.4

```python
gray_pollen = img_as_float(io.imread('rice.jpg', as_gray=True))
threshold = 0.4
binary_pollen = gray_pollen.copy()
binary_pollen[gray_pollen>threshold] = 1.
binary_pollen[gray_pollen<threshold] = 0.
plt.imshow(binary_pollen, cmap='gray', vmax=1., vmin=0.)
plt.show()
```
