```python
leaves = io.imread('leaves.jpg')
gray_leaves = io.imread('leaves.jpg', as_gray=True)
```

```python
plt.imshow(leaves)
plt.show()
plt.imshow(gray_leaves, cmap='gray')
plt.show()
```

```python
def increase_contrast(image):
    threshold = 0.15
    mod_image = image/image.max().copy()
    x1, y1 = np.where(mod_image>=threshold)
    x2, y2 = np.where(mod_image<threshold)
    mod_image[x1, y1] = mod_image[x1, y1] + 0.3
    mod_image[x2, y2] = mod_image[x2, y2] - 0.3
    return mod_image

leaves_gray = io.imread('leaves.jpg', as_gray=True)
plt.imshow(leaves_gray, cmap='gray')
plt.show()
```

```python
leaves_contrast= increase_contrast(leaves_gray)
```

```python
plt.imshow(leaves_contrast, cmap='gray', vmin=0.0, vmax=1.0)
plt.show()
```

```python
def decrease_contrast(image):
    threshold = 0.01
    mod_image = image/image.max().copy()
    x1, y1 = np.where(mod_image>=threshold)
    x2, y2 = np.where(mod_image<threshold)
    mod_image[x1, y1] = mod_image[x1, y1] + 0.3
    mod_image[x2, y2] = mod_image[x2, y2] - 0.3
    return mod_image

leaves_gray = io.imread('leaves.jpg', as_gray=True)
plt.imshow(leaves_gray, cmap='gray')
plt.show()
```

```python
leaves_contrast= decrease_contrast(leaves_gray)

plt.imshow(leaves_contrast, cmap='gray', vmin=0.0, vmax=1.0)
plt.show()
```
