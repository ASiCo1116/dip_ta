```python
import matplotlib.pyplot as plt
import numpy as np
from skimage import filters, io

rice = io.imread('rice.jpg', as_gray=True)
plt.imshow(rice, cmap='gray')
plt.show()
```

```python
from skimage.segmentation import flood, flood_fill

block_size = 155
local_thresh = filters.threshold_local(rice, block_size, offset=-20)
local_binary = rice > local_thresh
plt.imshow(local_binary, cmap='gray')
plt.show()

local_image = local_binary.copy()

local_image = local_image.astype(np.uint8)
local_image[local_image == 1] = 255

grain_count = 0
rows, cols = local_image.shape

for i in range(rows):
    for j in range(cols):
        pixel = local_image[i, j]
        if pixel == 255:
            grain_count += 1
            flood_fill(local_image, (i, j), 127, in_place=True)

print("Number of grains", grain_count)
plt.imshow(local_image, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
from skimage import measure

contour_image = local_binary.copy()
contour_image = contour_image.astype(float)

contours = measure.find_contours(contour_image, fully_connected='high')

fig, ax = plt.subplots()
ax.imshow(contour_image, cmap='gray')

for contour in contours:
    ax.plot(contour[:, 1], contour[:, 0], linewidth=2, color='red')

print("Number of grains", len(contours))

ax.axis('image')
# ax.set_xticks([])
# ax.set_yticks([])
plt.show()
```

```python
import numpy as np
from skimage.color import rgb2hsv

leaves= io.imread('leaves3.JPG')
plt.imshow(leaves)
plt.show()

leaves_hsv = rgb2hsv(leaves)
```

```python
leaves_binary = np.zeros(leaves.shape)
leaves_binary[np.logical_and(leaves_hsv[:, :, 0] > 0.17, leaves_hsv[:, :, 0] < 0.26)] = 1

plt.imshow(leaves_binary)
plt.show()
leaves_binary_gray = leaves_binary[:, :, 0]
```

```python
from skimage import morphology, img_as_float
```

```python
leaves_open = morphology.opening(leaves_binary_gray, morphology.square(59))
plt.imshow(leaves_open, cmap='gray')
plt.show()
```

```python
leaves_erosion = morphology.erosion(leaves_open, morphology.square(69))
plt.imshow(leaves_erosion, cmap='gray')
plt.show()
```

```python
leaves_dilation = morphology.dilation(leaves_erosion, morphology.square(3))
plt.imshow(leaves_dilation, cmap='gray')
plt.show()
```

```python
leaves_close = morphology.closing(leaves_dilation, morphology.square(59))
plt.imshow(leaves_close, cmap='gray')
plt.show()
```

```python
leaves_count_image = leaves_close.copy()

leaves_count_image = leaves_count_image.astype(np.uint8)
leaves_count_image[leaves_count_image == 1.] = 255

leaf_count = 0
rows, cols = leaves_count_image.shape

for i in range(rows):
    for j in range(cols):
        pixel = leaves_count_image[i, j]
        if pixel == 255:
            leaf_count += 1
            flood_fill(leaves_count_image, (i, j), 127, in_place=True)

print("Number of leaves", leaf_count)
plt.imshow(leaves_count_image, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
contour_image = leaves_close.copy()
contour_image = contour_image.astype(float)

contours = measure.find_contours(contour_image, fully_connected='high', positive_orientation='low')

fig, ax = plt.subplots()
ax.imshow(contour_image, cmap='gray')

for contour in contours:
    ax.plot(contour[:, 1], contour[:, 0], linewidth=1, color='red')

print("Number of leaves", len(contours))

ax.axis('image')
# ax.set_xticks([])
# ax.set_yticks([])
plt.show()
```
