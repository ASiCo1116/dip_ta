```python
from skimage import io, data, img_as_float #import the io, data module from scikit-image
import matplotlib.pyplot as plt #import the pyplot module to load and save the image
import numpy as np
```

```python
# (1)
rice = io.imread('rice.jpg', as_gray=True)
plt.imshow(rice, cmap='gray')
plt.show()

# (2)
plt.imshow(img_as_float(rice)*1.5, cmap='gray', vmin=0.0, vmax=1.0)
plt.show()

# (3)
plt.imshow(img_as_float(rice)/1.5, cmap='gray', vmin=0.0, vmax=1.0)
plt.show()
```

```python
# (1)
highway1 = io.imread('highway1.jpg')
highway2 = io.imread('highway2.jpg')

highway1_float = img_as_float(highway1)
plt.imshow(highway1_float, cmap='gray')
plt.show()

highway2_float = img_as_float(highway2)
plt.imshow(highway2_float, cmap='gray')
plt.show()

subtraction = highway2_float - highway1_float

plt.imshow(subtraction, cmap='gray')
plt.show()

# (2)
plt.imshow(highway2-highway1, cmap='gray')
plt.show()
print(highway2)
print(highway1)
print(highway2-highway1)
```

```python
# (1)
def decrease_contrast(image):
    threshold = 0.15
    mod_image = image/image.max().copy()
    x1, y1 = np.where(mod_image>=threshold)
    x2, y2 = np.where(mod_image<threshold)
    mod_image[x1, y1] = mod_image[x1, y1] - 0.15
    mod_image[x2, y2] = mod_image[x2, y2] + 0.15
    return mod_image

leaves_gray = io.imread('leaves.jpg', as_gray=True)
plt.imshow(leaves_gray, cmap='gray')
plt.show()

leaves_contrast = decrease_contrast(leaves_gray)
plt.imshow(leaves_contrast, cmap='gray', vmin=0.0, vmax=1.0)
plt.show()
```
