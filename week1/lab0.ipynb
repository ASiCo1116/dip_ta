# Load an image
### (1) Import the required library, i.e. skimage, matplotlit.pyplot
### (2) Use the 'imread' in skimage.io to read the sample image
### (3) Use 'imshow', 'show' in matplotlib.pyplot to show the image on the screen
### (4) Show the built-in image 'coins' from skimage.data

```python
from skimage import io, data, img_as_float #import the io, data module from scikit-image
import matplotlib.pyplot as plt #import the pyplot module to load and save the image
import numpy as np
```

```python
deer = io.imread('deer.jpg') #read the deer image
gray_deer = io.imread('deer.jpg', as_gray=True) #read the deer image as grayscale

plt.imshow(deer) #display the data as an image
plt.show() #show the image on the screen
plt.imshow(gray_deer)
plt.show()
plt.imshow(gray_deer, cmap='gray') #show the image in the assigned color map
plt.show()
```

# Save an image
### (1) Use the imsave in skimage.io to save the image

```python
io.imsave('deer1.jpg', deer)
io.imsave('deer2.png', gray_deer)
```

# Image addition

```python
coins = data.coins()
plt.imshow(coins, cmap='gray')
plt.show()

coins_float = img_as_float(coins)

plt.imshow(coins_float, cmap='gray', vmax=1.)
plt.show()

coins_float_02 = coins_float + 0.2

plt.imshow(coins_float_02, cmap='gray', vmax=1.)
plt.show()

coins_float_03 = coins_float - 0.3

plt.imshow(coins_float_03, cmap='gray', vmax=1.)
plt.show()
```

# Image substraction

```python
highway1 = io.imread('highway1.jpg')
highway2 = io.imread('highway2.jpg')

highway1 = highway1.astype(float)
plt.imshow(highway1, cmap='gray')
plt.show()

highway2 = highway2.astype(float)
plt.imshow(highway2, cmap='gray')
plt.show()

substraction = highway2 - highway1

plt.imshow(substraction, cmap='gray')
plt.show()
```

# Image multiplication

```python
camera = data.camera()
camera_float = img_as_float(camera) # read the image in the float type

plt.imshow(camera, cmap='gray')
plt.show()

camera_15 = (camera_float * 1.5)
plt.imshow(camera_15, cmap='gray', vmax=1.)
plt.show()

camera_075 = (camera_float * 0.75)
plt.imshow(camera_075, cmap='gray', vmax=1.)
plt.show()
```

# Image contrast

```python
def increase_contrast(image):
    threshold = 0.6
    w, h = image.shape
    mod_image = image/image.max().copy()
    x1, y1 = np.where(mod_image>=threshold)
    x2, y2 = np.where(mod_image<threshold)
    mod_image[x1, y1] = mod_image[x1, y1] + 0.2
    mod_image[x2, y2] = mod_image[x2, y2] - 0.2
    return mod_image

rice = io.imread('rice.jpg')
plt.imshow(rice, cmap='gray')
plt.show()

high_rice = increase_contrast(rice)
plt.imshow(high_rice, cmap='gray', vmax=1., vmin=0.)
plt.show()
```

# Binarization

```python
threshold = 0.5
gray_deer = io.imread('deer.jpg', as_gray=True)
binary_deer = gray_deer.copy()
binary_deer[gray_deer>threshold] = 1.
binary_deer[gray_deer<threshold] = 0.
plt.imshow(binary_deer, cmap='gray')
plt.show()
```
